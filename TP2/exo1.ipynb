```python
from sklearn.cluster import AgglomerativeClustering
from scipy.cluster.hierarchy import dendrogram, linkage
import numpy as np
from matplotlib import pyplot as plt
```

```python
# Données d'exemple
X = np.array([[12, 39],
              [20, 36],
              [28, 30],
              [18, 52],
              [29, 54],
              [33, 46],
              [24, 55],
              [45, 59],
              [45, 63],
              [52, 70],
              [51, 66],
              [52, 63],
              [55, 58],
              [53, 23],
              [55, 14],
              [61, 8],
              [64, 19],
              [69, 7],
              [72, 24]])
```

```python
# Créer le modèle de clustering avec les paramètres désirés
single_linkage = AgglomerativeClustering(n_clusters=3, linkage='single')
complete_linkage = AgglomerativeClustering(n_clusters=3, linkage='complete')
average_linkage = AgglomerativeClustering(n_clusters=3, linkage='average')
```

```python
# Entraîner le modèle
single_linkage.fit(X)
complete_linkage.fit(X)
average_linkage.fit(X)

# Obtenir les labels des clusters attribués à chaque point
single_linkage_labels = single_linkage.labels_
complete_linkage_labels = complete_linkage.labels_
average_linkage_labels = average_linkage.labels_

print("Labels des clusters :")
print(single_linkage_labels)
print(complete_linkage_labels)
print(average_linkage_labels)
```

```python
# Calculer la matrice de liens
# Créer le dendrogramme
single_Z = linkage(X, method='single')
dendrogram(single_Z)
plt.show()
```

```python
complete_Z = linkage(X, method='complete')
dendrogram(complete_Z)
plt.show()
```

```python
average_Z = linkage(X, method='average')
dendrogram(average_Z)
plt.show()
```
